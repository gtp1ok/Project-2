```python
csv_file = "Resources/Electric_Vehicle_Title_and_Registration_Activity.csv"
reg_df = pd.read_csv(csv_file)
reg_df.head()
reg_df.columns
```

```python
reg_df =reg_df[(reg_df['Model Year'] == 2020)]
reg_df
```

```python
reg_data_df = reg_df[['VIN (1-10)', 'Model Year', 'Make', 'Model', 'County', 'City', 'Postal Code']].copy()
reg_data_df
```

```python
csv_file = "Resources/Electric_Vehicle_Population_Size_History_By_County.csv"
pop_df = pd.read_csv(csv_file)
pop_df.head()
```

```python
# pop_df.drop(pop_df.index[pop_df['State'] != 'WA'], inplace=True)
# pop_df[pop_df['State'] != 'WA']
wadp = pop_df.loc[pop_df["State"] == 'WA']
wadp = pop_df.loc[pop_df["County"] == 'Adams']
wadp
```

```python
pop_data_df = wadp[['County', 'Battery Electric Vehicles (BEVs)', 'Plug-In Hybrid Electric Vehicles (PHEVs)', 'Non-Electric Vehicle Total']].copy()
pop_data_df
```

```python
# merged_df=pd.merge(clean_df,df,left_on='DOL Vehicle ID', right_on='dol_vehicle_id',how='left')
merge_df = pd.merge( pop_data_df,reg_data_df, on='County', how='left')
merge_df.rename(columns = {'Postal Code':'ZIP'}, inplace = True)
merge_df
```

```python
merge2_df = merge_df.drop(['VIN (1-10)', 'Model Year', 'Make', 'Model'], axis=1)
merge2_df
```
